```python
from bokeh.plotting import output_notebook, figure, show
from bokeh.models import HoverTool, tools
from bokeh.models import Span, Label, Arrow, NormalHead, ColumnDataSource
from bokeh.layouts import column
from astropy.io import fits
import numpy as np
import glob
from bokeh.models import HoverTool, tools, ColumnDataSource, CustomJS, Slider, BoxAnnotation, Range1d
from bokeh.models import BoxZoomTool

from astropy.io import fits
import os
from stsci.tools import capable
capable.OF_GRAPHICS = False
from pyraf import iraf
#Tool to get wavelength
hover2 = HoverTool(
        tooltips=[
            ("(x,y)", "($x{1}, $y)"),
        ]
    )


output_notebook()
```

# Sky substracted 

```python
noskylist  = glob.glob('xs*ms.fits')

rowplot = []

for sci in noskylist:
    sourcename = sci
    #### sourcename = noskylist[0]
    #Not sure if I need to do this
    if os.path.exists(sourcename+'.dispcor.fits'):
        os.remove(sourcename+'.dispcor.fits')
    iraf.onedspec.dispcor(sourcename+'[SCI]',sourcename+'.dispcor.fits')
    iraf.wspectext(sourcename+'.dispcor.fits[*,1,1]',sourcename+'.txt',header='no')

    x=[]
    y=[]
    with open(sourcename+'.txt') as f:
        for lines in f:
            x.append(float(lines.split()[0]))
            y.append(float(lines.split()[1]))


    source = ColumnDataSource(data=dict(x=x,y=y))
    hover = HoverTool(
            tooltips=[
                #("index", "$index"),
                ("(x,y)", "($x{1.11}, $y)"),
            ]
        )

    plot = figure(x_axis_label='Angstrom', y_axis_label='',title=sci,plot_width=800, plot_height=700,background_fill_color="#fafafa")
    plot.add_tools(hover)
    plot.y_range = Range1d(-10,10000)
    plot.x_range = Range1d(4500,10000)

    #plot.add_tools(tools.ResizeTool())
    #Eraaseplot.line(xlist,secondstar)
    #plot.toolbar.active_drag = BoxZoomTool()
    plot.line('x','y',source=source)
    rowplot.append(plot)
    
```

```python
show(column(rowplot))
```
